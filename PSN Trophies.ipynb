```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

import time
import datetime
import requests
```

```python
def get_trophy_title(r):
    return r.find(class_="title").contents[0]
```

```python
def get_trophy_description(r):
    return r.find(attrs={'style':'width: 100%;'}).contents[-1].split("\r")[0]
```

```python
def get_game(r):
    return r.find(class_='game')["title"]
```

```python
def get_trophy_type(r):
    return r.findAll(class_='separator left')[-1].find('img')["title"]
```

```python
def get_date(r):
    s = str(r.find(class_='typo-top-date').contents[0])[6:-7]
    return s.split('<')[0] + s.split('>')[-1]
```

```python
def get_hour(r):
    return r.find(class_="typo-bottom-date").contents[0].contents[0]
```

```python
def get_trophy_no(r):
    return int(r.find(attrs={'style':'text-align: right;'}).contents[1].contents[0][1:].replace(',',''))
```

```python
def get_achievers(r):
    return int(r.findAll(class_="typo-top")[0].contents[0].replace(',',''))
```

```python
def get_owners(r):
    return int(r.findAll(class_="typo-top")[1].contents[0].replace(',',''))
```

```python
def get_rarity(r):
    return r.findAll(class_="separator left")[3].find('nobr').contents[0]
```

```python
def get_game_link(r):
    return r.findAll('a')[0]['href']
```

```python
def get_trophy_img(r):
    return r.find(class_='trophy')["src"]
```

```python
trophy_nos = []
trophy_games = []
trophy_titles = []
trophy_descriptions = []
trophy_types = []
trophy_dates = []
trophy_hours = []
trophy_achievers = []
trophy_owners = []
trophy_rarities = []
game_links = []
trophy_imgs = []
```

```python
def append_trophy_attrs(rows):
    for r in rows:
        trophy_nos.append(get_trophy_no(r))
        trophy_games.append(get_game(r))
        trophy_titles.append(get_trophy_title(r))
        trophy_descriptions.append(get_trophy_description(r))
        trophy_types.append(get_trophy_type(r))
        trophy_dates.append(get_date(r))
        trophy_hours.append(get_hour(r))
        trophy_achievers.append(get_achievers(r))
        trophy_owners.append(get_owners(r))
        trophy_rarities.append(get_rarity(r))
        game_links.append(get_game_link(r))
        trophy_imgs.append(get_trophy_img(r))
```

```python
for i in range(1,23):
    resp = requests.get("https://psnprofiles.com/chilaquiles4life/log?page=" + str(i))
    soup = BeautifulSoup(resp.content, 'html.parser')
    table = soup.find('table', attrs={'class':'zebra'})
    append_trophy_attrs(table.find_all('tr'))
```

```python
trophies = pd.DataFrame({
    'trophy_no':trophy_nos,
    'game':trophy_games,
    'title':trophy_titles,
    'description':trophy_descriptions,
    'trophy_type':trophy_types,
    'date':trophy_dates,
    'hour':trophy_hours,
    'achievers':trophy_achievers,
    'owners':trophy_owners,
    'rarity':trophy_rarities,
    'game_link':game_links,
    'trophy_img':trophy_imgs
})
```

```python
def get_total_trophies(t):
    return int(t.find('span', class_='small-info').contents[3].contents[0])
```

```python
def get_completion_pct(t):
    return int(t.find(class_='progress-bar').find('span').contents[0].replace('%',''))
```

```python
def get_mark(t):
    return t.find('center').find('span').contents[0]
```

```python
#table = soup.findAll('table', attrs={'class':'zebra'})[1]
```

```python
def get_platform(t):
    return t.find(class_='platforms').contents[0].contents[0]
```

```python
#table = soup.findAll('table', attrs={'class':'zebra'})[2]
```

```python
def get_publisher(t):
    return t.findAll('tr')[0].find('a').contents[0]
```

```python
def get_genre(t):
    return t.findAll('tr')[2].find('a').contents[0].split(',')[0]
```

```python
links = []
total_trophies = []
completion_pcts = []
marks = []
platforms = []
publishers = []
genres = []
game_imgs = []
```

```python
for l in trophies["game_link"].drop_duplicates():
    resp = requests.get("https://psnprofiles.com" + l)
    soup = BeautifulSoup(resp.content, 'html.parser')

    links.append(l)
    table = soup.find('table', attrs={'class':'box'})
    total_trophies.append(get_total_trophies(table))
    completion_pcts.append(get_completion_pct(table))
    marks.append(get_mark(table))
    table = soup.findAll('table', attrs={'class':'zebra'})
    table = list(filter(lambda x: len(str(x).split('tag platform'))>1 and len(str(x).split(' zebra '))==1
                   , table))[0]
    platforms.append(get_platform(table))
    table = soup.findAll('table', attrs={'class':'zebra'})
    table = list(filter(lambda x: len(str(x).split('Publisher'))>1
                   , table))[0]
    publishers.append(get_publisher(table))
    genres.append(get_genre(table))
    game_imgs.append(soup.find(property='og:image')["content"])
```

```python
games = pd.DataFrame({
    'game_link':links,
    'total_trophies':total_trophies,
    'completion_pct':completion_pcts,
    'mark':marks,
    'platform':platforms,
    'publisher':publishers,
    'genre':genres,
    'game_img':game_imgs
})
```

```python
games["genre"] = games["genre"].apply(lambda x: 'Platform' if x < ' ' else x)
```

```python
games["genre"] = games.apply(lambda x: 'Visual Novel' if x.publisher == 'Spike ChunSoft' else x.genre, axis=1)
```

```python
trophies = trophies.merge(games, how='inner', on='game_link')
```

```python
trophies["rarity_pct"] = trophies.apply(lambda x: float(x.achievers)/x.owners, axis=1)
```

```python
trophies["timestamp"] = trophies.apply(lambda t: datetime.datetime.strptime(
        t.date + " " + t.hour, "%d %b %Y %I:%M:%S %p"
    ) - datetime.timedelta(hours=3), axis=1)
```

```python
trophies["weekday"] = trophies.timestamp.apply(lambda x: x.strftime('%A'))
```
